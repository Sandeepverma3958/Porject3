```python
import os
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from PIL import Image
```

```python
image = r"/content/_100058428_mediaitem100058424.jpg.webp"
```

```python
image = Image.open(image)
image
```

```python
image_RGB = image.convert('RGB')
image_RGB
```

```python
image_grayscale = image_RGB.convert('L')
image_grayscale
```

```python
width ,height = image_grayscale.size
print('The widht of the image :',width)
print('The height of the image',height)
```

```python
image_grayscale_resize = image_grayscale.resize((28,28))
```

```python
widht1,height1 = image_grayscale_resize.size
print('The widht of the image',widht1)
print('The height of the image',height1)
```

```python
image_grayscale_resize_array = np.array(image_grayscale_resize)
image_grayscale_resize_array
```

```python
print('Dimension of the pixelated array',image_grayscale_resize_array.ndim)
```

```python
print('Shape of the image',image_grayscale_resize_array.shape)
```

```python
plt.figure(figsize=(4,4))
plt.imshow(image_grayscale_resize_array,cmap = 'gray')
plt.colorbar()
plt.show()
```
